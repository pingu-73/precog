```python
nlp = spacy.load('en_core_web_lg');
```

```python
doc = nlp(text)
l = []
# for token in doc:
#     print(token.text)
for sent in doc.sents:
    # print(sent.text)
    sent_t = sent.text
    l.append(simple_preprocess(sent_t))
```

```python
len(l)
```

```python
model = Word2Vec(
    window = 10,
    min_count = 1,
    workers = 3
)
model.build_vocab(l)
model.train(l, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.wv.similarity('old', 'new')
```
